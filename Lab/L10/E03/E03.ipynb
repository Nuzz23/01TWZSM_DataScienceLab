```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from seaborn import heatmap


mat = np.zeros((100, 2))
mat[:, 0] = list(range(100))
mat[:, 1] = mat[:, 0]*2   # y = 2x


df = pd.DataFrame(mat, columns=['x', 'y'])

plt.plot(df.iloc[:, 0], df.iloc[:, 1])
plt.tight_layout()
```

```python
heatmap(df.corr(), annot=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from seaborn import heatmap
from math import sin


mat = np.zeros((100, 2))
mat[:, 0] = list(range(100))
mat[:, 1] = list(map(lambda x: sin(x), mat[:, 0]))  # y = sin(x), y=ax+b 

df = pd.DataFrame(mat, columns=['x', 'y'])

plt.plot(df.iloc[:, 0], df.iloc[:, 1])
```

```python
heatmap(df.corr(), annot=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score, explained_variance_score


mat = np.zeros((100, 2))
mat[:, 0] = list(range(100))
mat[:, 1] = mat[:, 0] + np.random.normal(0, scale=3, size=100)

df = pd.DataFrame(mat, columns=['x', 'y'])

plt.plot(df.iloc[:, 0], df.iloc[:, 1])
plt.plot(list(range(100)), list(range(100)))
plt.grid(True)

# y = b0 + b1*x + eps
# b0 = 0
# b1 = 1
# 
r2_score(list(range(100)), df['y'])
explained_variance_score(list(range(100)), df['y'])

# 1 - eps 
```
