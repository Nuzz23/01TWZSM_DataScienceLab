### HOW TO CREATE A FIGURE

we create a sub plot and specify the size of the figure

we get the figure object and the axis object
fig, ax = plt.subplots(figsize(1,2))

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(5,3))
print(type(fig))
print(type(ax))
ax.plot([1,2,3], [4,5,6])
ax.plot([1,2,3], [1,4,9])


fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(5,4))
# rearranges the layout to have no overlap
plt.tight_layout()
print(type(ax)) # becomes a numpy array

x = np.linspace(start=-10, stop=10, num=100)

for i in range(3):
    for j in range(2):
        ax[i,j].plot(x, x**(i*2*(j+1))) # we can now plot on just one of the subplot


plt.show() # used only if you use python


```
