```python
from sklearn.feature_extraction.text import TfidfTransformer

from sklearn.decomposition import PCA


```

```python
import pandas as pd

with open('dati.txt', 'r', encoding='UTF-8') as fp:
    df = pd.DataFrame([[' '.join(line.split(',')[:-1]), line.strip().split(',')[-1]] for line in fp], columns=['review', 'label'])

df = df.iloc[1:, :]    

from sklearn.model_selection import train_test_split


X = df.loc[:, 'review']
y = df.loc[:, 'label']
display(df)
# Stratify to have the same distribution and no different values
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size=0.25)


# Count Ã¬Vectorize to split the word in words and count
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

pipe = make_pipeline(CountVectorizer(stop_words='english'), DecisionTreeClassifier(max_depth=10)).fit(xTrain, yTrain)


yPred = pipe.predict(xTest)

from sklearn.metrics import accuracy_score
accuracy_score(yTest, yPred)*100

```
